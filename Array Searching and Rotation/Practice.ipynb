## **Linear Search**

```python
def linear_search(arr,target):
    found = False
    for i in range(0,len(arr)):
        if arr[i] == target:
            found = True 
            return f"Element found at Index = {i}"
    if not found:
        return "Element not found"
print(linear_search([5,4,3,1,2],5))
```

## **Binary Search**

```python
def binary_search(arr,target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = low + (high - low) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1 

print(binary_search([1,3,5,7,11],7))
```

## **Rotate Arrays**

```python
def rotate_arrays(arr,k,is_left):
    k = k % len(arr)
    if is_left: ## left rotation
        arr = arr[k:] + arr[:k]
    else: ## right rotation
        arr = arr[-k:] + arr[:-k]
    return arr
    
print(rotate_arrays([1,2,3,4,5],1,True))
print(rotate_arrays([1,2,3,4,5],2,False))
```
