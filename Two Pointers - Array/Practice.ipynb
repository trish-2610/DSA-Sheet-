## **Remove Duplicates from sorted array**

```python
def remove_duplicates_sorted(arr):
    i = 0
    for j in range(1,len(arr)):
        if arr[j] != arr[i]:
            i += 1
            arr[i] = arr[j]
    return arr[:i+1]

remove_duplicates_sorted([1,1,2,2,2,3,3,4,5])
```

## **Even-Odd Partitioning**

```python
def even_odd_partition(arr):
    left = 0
    right = len(arr) - 1
    while left < right:
        while arr[left] % 2 == 0 and left < right: ## tracks even on left
            left += 1
        while arr[right] % 2 == 1 and left < right: ## tracks odd on right
            right -= 1
        if left < right:
            ## Swap - if both loops doesn't work
            arr[left],arr[right] = arr[right],arr[left]
    return arr

even_odd_partition([12,17,70,15,22,65,21,90])
```

## **Positive-Negative Partitioning**

```python
def pos_neg_partition(arr):
    left = 0 
    right = len(arr) - 1
    while left < right:
        while arr[left] < 0 and left < right: ## tracks negative 
            left += 1 
        while arr[right] >= 0 and left < right: ## tracks positive
            right -= 1
        if left < right:
            ## Swap - if both loop fails
            arr[left],arr[right] = arr[right],arr[left]
    return arr 

pos_neg_partition([-1,2,3,4,-2,-6,1])
```

## **Two Sum-II**

```python
def two_sum_II(arr,target):
    left = 0 
    right = len(arr) - 1
    while left < right:
        cur_sum = arr[left] + arr[right]
        if cur_sum == target:
            return [left+1,right+1] ## indexing from 1 not 0
        elif cur_sum < target:
            left += 1
        else:
            right -= 1
    return -1

two_sum_II([1,2,3,4,6,8,9],4)
```
