## **Implementation of Stack**

```python
## using list 
stack = [10,20,30,40,50]
stack.append(60)
stack.append(70)
stack.pop()
stack.append(80)
stack.pop()
print(stack)
```

```python
## using deque
from collections import deque
stack = deque()
stack.append(10)
stack.append(20)
stack.append(30)
stack.pop()
stack.pop()
print(stack)
print("TOS =",stack[-1])
```

```python
## using custom class 
class Stack:
    def __init__(self):
        self.container = []
    def insert(self,item):
        self.container.append(item)
    def is_empty(self):
        return len(self.container) == 0
    def pop(self):
        return self.container.pop() if not self.is_empty() else None
    def peek(self):
        return self.container[-1] if not self.is_empty() else None
    def size(self):
        return len(self.container)

stack = Stack()
stack.insert(10)
```

```python
## Postfix Expression Evaluation
def evaluate_expression(expr):
    stack = []
    for char in expr.split():
        if char.isdigit():
            stack.append(int(char))
        else:
            b = stack.pop()
            a = stack.pop()
            if char == "+" : stack.append(a+b)
            elif char == "-" : stack.append(a-b)
            elif char == "*" : stack.append(a*b)
            elif char == "/" : stack.append(int(a/b))
    return stack.pop()

expr = "5 6 2 + * 12 4 / -"
print(evaluate_expression(expr))
```

```python
## Valid Parentheses
def valid_parentheses(expr):
    stack = []
    pairs = {
             "}":"{",
             "]":"[",
             ")":"("
            }
    for char in expr:
        if char in "{[(":
            stack.append(char)
        elif char in "}])":
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()
    return not stack
print(valid_parentheses("(([]))"))
```
